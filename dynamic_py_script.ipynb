```python
#the function loops through each column of the DF and checks if it contains text or numerical values 
# ----------------------
# for text columns, it fills any missing values with the most frequent value (mode)
#-----------------------
# for numerical columns, it fills missing values with the mean of the available values 
#-----------------------
# Next, the function removes duplicate rows from the DF
#-----------------------
# Finally, the cleaned DF is returned 

def clean_data(data):
    for column in data.columns:
        if data[column].dtype == 'object':
            data[column].fillna(data[column].mode()[0],
                                inplace = True)
        else:
            data[column].fillna(data[column].mean(),
                                inplace = True)
    data.drop_duplicates(inplace = True)
    return(data)

cleaned_data = clean_data(data)
print(cleaned_data)
```
